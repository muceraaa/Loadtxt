```python
# define the function f(t) = a*cos(2t*pi + b) + c
def f(t, a, b, c):
    return a*np.cos(2*t*np.pi + b) + c


data = pd.read_csv('weather_data_nyc_centralpark_2016(1).csv', usecols=['date', 'average temperature'],
                   parse_dates=['date'], index_col=['date'], squeeze=True)

keep = np.abs(data.values) < 90
filtered_data = data[keep]

# define the time as a fraction of the days since the first measurement
t = (filtered_data.index - filtered_data.index[0]).days

# define the parameters a, b, and c 
a = 1.0
b = 0.0
c = 0.0

f(t, a, b, c)
```

```python
from scipy.optimize import curve_fit
a_bounds = (np.inf, -np.inf)
b_bounds = (-np.inf, -np.inf)
c_bounds = (-np.inf, -np.inf)
popt, _= curve_fit(f, t, data, p0=[10, 0, 10], bounds=(a_bounds, b_bounds, c_bounds))
```
